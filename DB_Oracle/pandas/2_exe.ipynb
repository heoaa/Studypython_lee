# 함수

```python
import cx_Oracle as ora
import pandas as pd

# 오라클 연결/접속/커서
def setDsn() :
    return ora.makedsn('localhost', 1521, service_name = 'orcl')

def setConn(dsn) :
    return ora.connect('busan_06', 'dbdb', dsn)

def setCur(conn) :
    return conn.cursor()

# 오라클 커서 반납, 접속 끊기
def closeOracle(cur, conn) :
    cur.close()
    conn.close()

# 데이터 가져오기
def getFetchall(query, cur) :
    cur.execute(query)
    return cur.fetchall()

# 데이터프레임 만들기
def getCol(cur) :
    col = []
    for i in cur.description :
        col.append(i[0])
    return col

def getDF(row, col) :
    return pd.DataFrame(row, columns=col)

# 최종 호출함수
def getAll(query) :
    dsn = setDsn()
    conn = setConn(dsn)
    cur = setCur(conn)
    row = getFetchall(query, cur)
    col = getCol(cur)
    closeOracle(cur, conn)
    return getDF(row, col)
```

## 함수 사용

```python
query = """ select * from member """

getAll(query)
```
