## 전체 코드

```python
import cx_Oracle as ora
import pandas as pd

dsn = ora.makedsn('localhost', 1521, service_name = 'orcl')
conn = ora.connect(user = 'busan_06', password = 'dbdb', dsn = dsn, encoding = 'UTF-8')
cur = conn.cursor()

query = """ select * 
from cart """

cur.execute(query)
row = cur.fetchall()
colname = cur.description

cur.close()
conn.close()
```

# 데이터 정리

## 컬럼명 가져오기

```python
colname
```

```python
# 인덱스를 이용한 반복문
col = []
for i in range(len(colname)) :
    col.append(colname[i][0])
col
```

```python
# 값을 이용한 반복문
col = []
for tp in colname :
    col.append(tp[0])
col
```

## 데이터프레임 만들기

```python
col
```

```python
row
```

```python
pd.DataFrame(row, columns=col)
```
