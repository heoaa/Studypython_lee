```python
starbucks_map = folium.Map(
                            location = [37.573050, 126.979189],       # 최초에 보여줄 지도의 중앙점 좌표 [위도, 경도]
                            tiles = 'openstreetmap',                  # 지도 스타일 (openstreetmap : 일반적 도시 형태)
                            zoom_start = 50                           # 최초 보여줄 지도의 스케일
)
starbucks_map
```

```python
# 데이터프레임에 있는 위도/경도 값을 각각 추출하여 출력

for i in range(0,len(seoul_starbucks)):
    a = seoul_starbucks['위도'][i]
    b = seoul_starbucks['경도'][i]
    # print(a, b)
    
    # 지도에  표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location = [a, b],         # 위치
        fill_color = 'yellow',      # 채우기 색상
        fill_opacity = 1,          # 채우기 불투명도
        color = 'green',          # 테두리 색상
        weight = 1,                # 테두리 굵기
        radius = 5                 # 전체 원의 지름
    ).add_to(starbucks_map)
    
starbucks_map
```

```python
# 스타벅스 매장 타입별로 마커 다르게 표시하기

starbucks_map2 = folium.Map(location = [37.573050, 126.979189],
                            tiles = 'openstreetmap',
                            zoom_start = 11)

for i in range(0,len(seoul_starbucks)):
    a = seoul_starbucks['위도'][i]
    b = seoul_starbucks['경도'][i]
    store_type = seoul_starbucks.loc[i, '매장타입']
    
    # general  : gray, 3
    # generalDT : red, 5
    # reserve : blue, 5
    
    fillcolor = ''
    size = 1
    
    if store_type == 'general' :
        fillcolor = 'gray'
        size = 3
    elif store_type == 'generalDT' :
        fillcolor = 'red'
        size = 5
    elif store_type == 'reserve' :
        fillcolor = 'blue'
        size = 5
    
    # 지도에  표시할 마킹 모양 등 스타일 지정
    folium.CircleMarker(
        location = [a, b],         # 위치
        fill_color = fillcolor,      # 채우기 색상
        fill_opacity = 1,          # 채우기 불투명도
        fill = True,
        color = fillcolor,          # 테두리 색상
        weight = 1,                # 테두리 굵기
        radius = size                 # 전체 원의 지름
    ).add_to(starbucks_map2)
    
starbucks_map2
```
