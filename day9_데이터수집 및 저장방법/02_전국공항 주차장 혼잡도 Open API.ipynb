```python
'''
airportEng 공항명(영문)
airportKor 공항명(한글)
parkingAirportCodeName 주차장명
parkingCongestion 주차장 혼잡정도
parkingCongestionDegree 주차장 혼잡률
parkingOccupiedSpace 입고된 차량 수
parkingTotalSpace 전체 주차면 수
'''
```

```python
airporteng_list = soup.find_all('airporteng')
airportkor_list = soup.find_all('airportkor')
p_aircd_list = soup.find_all('parkingairportcodename')
p_con_list = soup.find_all('parkingcongestion')
p_con_deg_list = soup.find_all('parkingcongestiondegree')
p_sp_list = soup.find_all('parkingoccupiedspace')
p_to_sp_list = soup.find_all('parkingtotalspace')
```

```python
list_all = []

for i in range(len(airporteng_list)) :
    list_all.append([airporteng_list[i].text, airportkor_list[i].text, 
                     p_aircd_list[i].text, p_con_list[i].text,
                     p_con_deg_list[i].text, p_sp_list[i].text,
                     p_to_sp_list[i].text])

# list_all
    
col_nm = ['공항명(영문)', '공항명(한글)', '주차장명', '주차장 혼잡정도', 
          '주차장 혼잡률', '입고된 차량 수', ' 전체 주차면 수']

df_gmp = pd.DataFrame(list_all, columns = col_nm)
df_gmp
```

## 실시간 공항 데이터 한번에 불러오기(10개 공항)

```python
## Open API 가져오기
apikey = 'PVdeh44WlF%2BHj6bGzfI2BXUldlea7Ggo%2FjjXvhbHCMNNFotchb2byGTopt4uAFS07RQzqmeU3M2Y8HcMmlYW9w%3D%3D'
schAirportCode = ['GMP','PUS','CJU','TAE','KWJ','RSU','USN','KUV','WJU','CJJ']
numOfRows = 10
pageNo = 1

api = "http://openapi.airport.co.kr/service/rest/AirportParkingCongestion/airportParkingCongestionRT?schAirportCode={aircode}&serviceKey={key}&numOfRows={row}&pageNo={no}"

list_all = []

for i in range(len(schAirportCode)) :
    url = api.format(aircode = schAirportCode[i], 
                     key = apikey, 
                     row = numOfRows, 
                     no = pageNo)

    response = requests.get(url)
    # print(response.text)

## html 형태로 의미 부여
    re = response.text
    soup = BeautifulSoup(re, 'html.parser')


    airporteng_list = soup.find_all('airporteng')
    airportkor_list = soup.find_all('airportkor')
    p_aircd_list = soup.find_all('parkingairportcodename')
    p_con_list = soup.find_all('parkingcongestion')
    p_con_deg_list = soup.find_all('parkingcongestiondegree')
    p_sp_list = soup.find_all('parkingoccupiedspace')
    p_to_sp_list = soup.find_all('parkingtotalspace')
    sys_date_list = soup.find_all('sysgetdate')
    sys_time_list = soup.find_all('sysgettime')


    for j in range(len(airporteng_list)) :
        list_all.append([airporteng_list[j].text, 
                         airportkor_list[j].text, 
                         p_aircd_list[j].text, 
                         p_con_list[j].text,
                         p_con_deg_list[j].text, 
                         p_sp_list[j].text,
                         p_to_sp_list[j].text,
                         sys_date_list[j].text, 
                         sys_time_list[j].text])


col_nm = ['공항명(영문)', '공항명(한글)', '주차장명', '주차장 혼잡정도', 
          '주차장 혼잡률', '입고된 차량 수', ' 전체 주차면 수',
          '업데이트 일자', '업데이트 시간']

df_air = pd.DataFrame(list_all, columns = col_nm)
df_air
```
