```python
import pandas as pd
```

# 전체파일 통합

```python
df = pd.DataFrame()
```

```python
for i in range(80):
    file_path = '../data/trfcard({})/trfcard.csv'.format(i)
    df_temp = pd.read_csv(file_path)
    df = pd.concat([df,df_temp], ignore_index=True)
```

```python
# 결측치 체크
df.info()
```

```python
# 이상치 체크
df.describe()
```

```python
# 중복값 체크
df.duplicated().sum()
```

# 컬럼명 한글로 변환

```python
file_path = '../data/trfcard(0)/trfcard_columns.xlsx'
col_kor_name = pd.read_excel(file_path,
                             skiprows=[0,1], usecols='B:C')
col_kor_name
```

```python
col_kor_name_list = col_kor_name.to_dict('records')
col_kor_name_list
```

```python
# 깨끗한 딕셔너리 변수 생성
new_dict = {}
```

```python
for i in col_kor_name_list:
    key,value = i.values()
    new_dict[key] = value
    
df_kor = df.rename(columns=new_dict)
df_kor
```

```python
df_kor.to_csv('../data/trfcard_result.csv', index=False)
```
