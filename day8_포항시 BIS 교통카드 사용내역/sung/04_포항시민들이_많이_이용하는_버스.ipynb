```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='Malgun Gothic')
```

```python
file_path = '../data/trfcard_result.csv'
df_cp = pd.read_csv(file_path)
df_cp
```

```python
df = df_cp.copy()
df
```

```python
# 월, 일, 시 컬럼 추가
df['승차월'] = df['승차시각'].astype(str).str[4:6]
df['승차일'] = df['승차시각'].astype(str).str[6:8]
df['승차시'] = df['승차시각'].astype(str).str[8:10]
```

```python
df['노선명'].value_counts()
```

포항 시민들이 가장 많이 이용하는 버스번호 top5
105번, 107번, 102번, 101번, 200번 입니다.

```python
top_bus = ['105','107','102','101','200']
```

```python
df_top_bus_en = pd.DataFrame()
df_top_bus_ex = pd.DataFrame()
```

```python
for i in top_bus:
    df_en = df[df['노선명'] == i]['승차정류장'].value_counts()[0:1]
    df_top_bus_en = pd.concat([df_top_bus_en, df_en])
    
for i in top_bus:
    df_ex = df[df['노선명'] == i]['하차정류장'].value_counts()[0:1]
    df_top_bus_ex = pd.concat([df_top_bus_ex, df_ex])
```

```python
# 포항 시민들이 가장 많이 이용하는 버스번호 top5의 가장 많이 탑승하는 승차정류장
df_top_bus_en
```

```python
# 포항 시민들이 가장 많이 이용하는 버스번호 top5의 가장 많이 하차하는 하차정류장
df_top_bus_ex
```
