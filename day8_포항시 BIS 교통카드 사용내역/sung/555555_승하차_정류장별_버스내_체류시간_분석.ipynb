```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
plt.rc('font', family='Malgun Gothic')
```

```python
file_path = '/data/trfcard_result.csv'
df_cp = pd.read_csv(file_path)
df_cp
```

```python
df = df_cp.copy()
df
```

```python
df['승차시각'] = df['승차시각'].astype(str)
df['승차시각']
```

```python
df['승차시각'] = pd.to_datetime(df['승차시각'])
```

```python
df['하차시각'] = df['하차시각'].astype(str)
df['하차시각']
```

```python
df['하차시각'] = pd.to_datetime(df['하차시각'])
```

```python
df['체류시간'] = df['하차시각'] - df['승차시각']
df
```

```python
df.to_csv('C:/DEV/STUDY/StudyPython22.06/day8/data/trfcard_time.csv')
```

```python
# '구간' 컬럼 추가
df['구간'] = df['승차정류장'] + ' - ' + df['하차정류장']
df
```

# 승차정류장으로 그룹

```python
df_enter = df.groupby('승차정류장')['체류시간'].mean()
df_enter
```

```python
df_enter = df_enter.sort_values(ascending=False).head(30)
df_enter
```

```python
df_enter_in = df_enter.index
df_enter_in
```

```python
df_enter_re = pd.DataFrame()
```

```python
for i in df_enter_in:
    df_temp = df[df['승차정류장'] == i]
    df_enter_re = pd.concat([df_enter_re, df_temp], ignore_index=True)
```

```python
df_enter_re
```
