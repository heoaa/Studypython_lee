```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv('./data/df_new_final.csv')
df
```

## 기준월 및 기준일자별 버스 이용량 분석 비교

```python
pivot_month_day = df.pivot_table(values = '승객연령',
                             index = '승차일', columns='승차월',
                             aggfunc = 'count')
pivot_month_day = pivot_month_day.fillna(0)
pivot_month_day
```

```python
a = []
for i in range(0,32):
    b = i
    a.append(b)
    
print(a)
```

```python
plt.rcParams['figure.figsize'] = (20,10)

plt.rcParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_month_day.plot()

plt.title('기준월 및 기준일자별 버스 이용량')
plt.ylabel('버스 이용량')

plt.xticks(a)

plt.legend(loc='upper right')

plt.grid(True)

plt.show()
```

```python
'''
1) 전반적으로 3월보다 1,2월에 버스 이용량이 높은 것을 확인할 수 있다. 
   1,2월의 경우 방학이나 겨울휴가로 인해 3월보다 높게 나타난 것으로 추정된다.
   
2) 1월 24, 25일 버스 이용량을 살펴보면 급격히 하락한 것을 확인할 수 있다. 
   이는 설날 연휴 기간으로 포항 관광객이 줄어들며 하락한 것으로 보인다.
   설날 연휴가 끝난 후 다시 회복세를 보인 것도 확인할 수 있다.
   
3) 2월을 살펴보면 19일부터 월말까지 하락세를 보이고 있다.
'''
```

## 기준일 및 시간대별 버스 이용량 분석 비교

```python
pivot_hour_day = df.pivot_table(values = '승객연령',
                             index = '승차일', columns='승차시간대',
                             aggfunc = 'count')
pivot_hour_day = pivot_hour_day.fillna(0)
pivot_hour_day
```

```python
plt.rcParams['figure.figsize'] = (20,10)

plt.rcParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_hour_day.plot()

plt.title('기준일 및 시간대별 버스 이용량')
plt.ylabel('버스 이용량')

plt.xticks(a)

plt.legend(loc='upper right')

plt.grid(True)

plt.show()
```

```python
'''
전반적으로 오후, 오전, 저녁,아침,밤 순으로 버스이용량이 높은 것을 확인할 수 있다.

시간대와 상관없이 20일부터 설연휴인 25일까지 점점 하락하다 회복하는 추세를 보인다. 
'''
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(pivot_hour_day, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('기준일 및 시간대별 버스 이용량')
plt.show()
```

### 각 시간의 성격을 기준으로 시간대 만들기

```python
df['승차시'].unique()
# 
```

```python
# 승차시간대
df['승차시간대'] = 'h'

row_index = df.index[(df['승차시'] >= 0) & (df['승차시'] <= 5)]
df.loc[row_index, '승차시간대'] = '새벽(0~5시)'

row_index = df.index[(df['승차시'] >= 6) & (df['승차시'] <= 8)]
df.loc[row_index, '승차시간대'] = '아침(6~8시)'

row_index = df.index[(df['승차시'] >= 9) & (df['승차시'] <= 12)]
df.loc[row_index, '승차시간대'] = '오전(9~12시)'

row_index = df.index[(df['승차시'] >= 13) & (df['승차시'] <= 16)]
df.loc[row_index, '승차시간대'] = '오후(13~16시)'

row_index = df.index[(df['승차시'] >= 17) & (df['승차시'] <= 19)]
df.loc[row_index, '승차시간대'] = '저녁(17~19시)'

row_index = df.index[(df['승차시'] >= 20) & (df['승차시'] <= 23)]
df.loc[row_index, '승차시간대'] = '밤(20~23시)'

df
```

```python
pivot_hour_min_g = df.pivot_table(values = '승객연령',
                             index = '승차시간대', columns='승차분대',
                             aggfunc = 'count')
pivot_hour_min_g = pivot_hour_min_g.fillna(0)
pivot_hour_min_g
```

```python
plt.rcParams['figure.figsize'] = (20,10)

plt.rcParams['lines.linewidth'] = 2

plt.rc('font', family = 'Malgun Gothic', size = 10)

pivot_hour_min_g.plot()

plt.title('기준시간 및 시간(분)별 버스 이용량')
plt.ylabel('버스 이용량')

# plt.xticks(a)

plt.legend(loc='upper right')

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize = (16,10))
sns.heatmap(pivot_hour_min_g, annot=True, fmt = '.0f', cmap='rocket_r')
plt.title('기준시간 및 시간(분)별 버스 이용량')
plt.show()
```
