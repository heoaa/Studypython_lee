## 데이터 정제 및 전처리

```python
# 데이터 합치기
df_e = pd.DataFrame()

for i in range(0, 80) :
    bus = pd.read_csv('./data/trfcard({})/trfcard.csv'.format(i))
    df_e = pd.concat([df_e, bus], ignore_index=True)

# 컬럼명 변경
df = pd.DataFrame()
n_dict = {}

df_col = pd.read_excel('./data/trfcard(0)/trfcard_columns.xlsx', skiprows=[0,1], usecols='B:C')

for j in range(len(df_col)) :
    n_dict[df_e.columns[j]] = df_col['컬럼명 (한글)'][j]

df = df_e.rename(columns = n_dict)
df
```

## 필요한 컬럼 생성

```python
df[['승차시각','하차시각']]
```

```python
df['승차년'] = df['승차시각'].astype(str).str[:4].astype(int)
df['승차월'] = df['승차시각'].astype(str).str[4:6].astype(int)
df['승차일'] = df['승차시각'].astype(str).str[6:8].astype(int)
df['승차시'] = df['승차시각'].astype(str).str[8:10].astype(int)
df['승차분'] = df['승차시각'].astype(str).str[10:12].astype(int)
df['승차초'] = df['승차시각'].astype(str).str[12:].astype(int)
```

```python
# 승차분대
df['승차분대'] = 'm'

for i in range(0, 60, 10) : 
    if i == 0 :
        row_index = df.index[df['승차분'] < i+10]
        df.loc[row_index, '승차분대'] = '10분미만'

    if i >= 10 :
        row_index = df.index[(df['승차분'] >= i) & (df['승차분'] < i+10)]
        df.loc[row_index, '승차분대'] = '{}분대'.format(i)
        
df
```

```python
# 승차시간대
df['승차시간대'] = 'h'

row_index = df.index[(df['승차시'] >= 0) & (df['승차시'] <= 5)]
df.loc[row_index, '승차시간대'] = '새벽(0~5시)'

row_index = df.index[(df['승차시'] >= 6) & (df['승차시'] <= 8)]
df.loc[row_index, '승차시간대'] = '아침(6~8시)'

row_index = df.index[(df['승차시'] >= 9) & (df['승차시'] <= 12)]
df.loc[row_index, '승차시간대'] = '오전(9~12시)'

row_index = df.index[(df['승차시'] >= 13) & (df['승차시'] <= 16)]
df.loc[row_index, '승차시간대'] = '오후(13~16시)'

row_index = df.index[(df['승차시'] >= 17) & (df['승차시'] <= 19)]
df.loc[row_index, '승차시간대'] = '저녁(5~7시)'

row_index = df.index[(df['승차시'] >= 20) & (df['승차시'] <= 23)]
df.loc[row_index, '승차시간대'] = '밤(20~23시)'

df
```
